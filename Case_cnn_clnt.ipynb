```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.models import load_model
from keras.preprocessing import image

cnn_model = load_model('C:/Users/lms/Documents/Visual Studio 2022/Case_cnn.h5')
cnn_model.summary()
```

```python
dir = 'C:/Users/lms/Documents/testTen/test'

filenames = os.listdir(dir)
filenames
```

```python
fig = plt.figure(figsize=(16,10))
rows, cols = 4, 5
for i, fn in enumerate(filenames):
    path = dir + '/' + fn
    test_img = image.load_img(path, color_mode='grayscale', target_size=(150, 150), interpolation='bilinear')
    x = image.img_to_array(test_img)
    x = np.expand_dims(x, axis=0)
    images = np.vstack([x])

    classes = cnn_model.predict(images, batch_size=10)

    fig.add_subplot(rows, cols, i+1)
    if classes[0]==0:
        plt.title(fn + " is O")
        plt.axis('off')
        plt.imshow(test_img, cmap='gray')
        print(fn + " is O")

    else:
        plt.title(fn + " is X")
        plt.axis('off')
        plt.imshow(test_img, cmap='gray')
        print(fn + " is X")
plt.show();
```
